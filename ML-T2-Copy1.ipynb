```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
%matplotlib inline
```

```python
lr_model = LinearRegression(normalize=True)
```

```python
#creating random data
```

```python
x = np.random.rand(200,1)*20
y = 1.5+2.1*x+2.8*(x**2)+1.3*(x**3)+2.7*(x**4)
y = y + np.random.normal(scale=3000,size=y.shape)
y = y.reshape(-1, 1)
x = x.reshape(-1, 1)
grade=[1,2,4,8,16]
ErrorTrain=[]
ErrorTest=[]
```

```python
#Normalize data
```

```python
n = int(x.shape[0] * 0.8)
xtrain = x[:n].reshape(-1, 1)
ytrain = y[:n].reshape(-1, 1)
xtest = x[n:].reshape(-1, 1)
ytest = y[n:].reshape(-1, 1)
```

```python
#T=1
lr_model.fit(xtrain,ytrain)
xline = np.arange(0,20, 0.1)
xline = xline.reshape(-1, 1)

plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("T= 1")
plt.plot(xline,lr_model.predict(xline),color='r',label="Model")
plt.plot(xtrain,ytrain,'bo',label="Train")
plt.plot(xtest,ytest,'o',color='c',label="Test")
plt.legend()

ErrorTrain.append(np.mean((ytrain - lr_model.predict(xtrain)) ** 2) / 2)
ErrorTest.append(np.mean((ytest - lr_model.predict(xtest)) ** 2) / 2)

train_sizes , train_score, test_score = learning_curve(lr_model,xtrain,ytrain, cv=5)
plt.figure()
plt.xlabel("Train Size")
plt.ylabel("Score")
plt.title("Learning Curve :T = 1")
plt.plot(train_sizes, train_score.mean(axis=1),'o-',label="Train")
plt.plot(train_sizes, test_score.mean(axis=1),'o-',label="Test")
plt.legend()
```

```python
#T=2
x_train=xtrain
x_train=np.hstack((x_train,xtrain**2)) 
lr_model.fit(x_train,ytrain)

xline = np.arange(0,20, 0.1)
xline = xline.reshape(-1, 1)
x_line=xline
x_line=np.hstack((x_line,xline**2)) 
plt.figure()
plt.xlabel("X")
plt.ylabel("Y")
plt.title("T = 2")
plt.plot(xline,lr_model.predict(x_line),color='r',label="Model")
plt.plot(xtrain,ytrain,'bo',label="Train")
plt.plot(xtest,ytest,'o',color='#33caba',label="Test")
plt.legend()

ErrorTrain.append(np.mean((ytrain - lr_model.predict(x_train)) ** 2) / 2)
x_test=xtest
x_test=np.hstack((x_test,xtest**2)) 
ErrorTest.append(np.mean((ytest - lr_model.predict(x_test)) ** 2) / 2)

train_sizes ,train_score, test_score = learning_curve(lr_model,x_train,ytrain, cv=5)
plt.figure()
plt.xlabel("Train Size")
plt.ylabel("Score")
plt.title("Learning Curve : T= 2")
plt.plot(train_sizes, train_score.mean(axis=1),'o-',label="Train")
plt.plot(train_sizes, test_score.mean(axis=1),'o-',label="Test")
plt.legend()
```
